### 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no son válidas las evaluaciones realizadas por los miembros del mismo equipo.
### Tu código a continuación...

```python
# SE LLAMAN LAS TABLAS DE ASISTENCIAS Y EVALUACIONES
tabla0 = ContextoSql.table('asistencias')
tabla = ContextoSql.table('evaluacion')
tabla.show(3) 
```

```python
from pyspark.sql import functions as F
from pyspark.sql.functions import expr

#tabla0 = ContextoSql.table('asistencias')
tabla01 = tabla0.withColumn("EquipoPertenece",expr("substring(EquipoPertenece, 15, 100 )"))# SE ELIMINA 'Pertenezco a: ' EN LA TABLA DE ASISTENCIAS DE LA COLUMNA 'EquipoPertenece'
tabla02 = tabla01.groupBy('NombreUsuario').agg(F.first('EquipoPertenece'))# SE AGRUPA LA ANTERIOR POR 'NombreUsuario' y se le agrega 'EquipoPertenece'
tmp2 = tabla
#tmp2.show()
join = tabla.join(tabla02, tabla['NombreUsuario'] == tabla02.NombreUsuario, how='left')#SE HACE UN LEFTJOIN DE EVALUACIONES CON LA TABLA ANTERIOR POR 'NombreUsuario'
tmp1 = join

# SE ELIMINAN LAS FILAS EN DONDE COINCIDAN EL GRUPO A EVALUAR('EquipoEvaluar') CON EL GRUPO AL QUE PERTENECE('EquipoPertenece')
tabla4 = join.filter(join['EquipoEvaluar'] != join['first(EquipoPertenece)'])
print(tabla.count())
print(join.count())
```

```python
from pyspark.sql import functions as F

#SE AGRUPA POR EQUIPO Y SE AGREGAN EL PROMEDIO DE LAS CALIFICACIONES
tabla5 = tabla4.groupBy('EquipoEvaluar').agg(F.first('default.evaluacion.NombreUsuario'),F.avg('Introduccion'),F.avg('Equipo'),F.avg('Problema'),F.avg('Ventajas'),F.avg('Solucion'),F.avg('Producto'),F.avg('Traccion'),F.avg('Mercado'),F.avg('Competencia'),F.avg('Negocio'),F.avg('Inversion'),F.avg('Contacto'),F.avg('ExpoCoord'),F.avg('ExpoExp'),F.avg('ExpoDiapo'))
tabla6 = tabla5.filter("avg(Introduccion) IS NOT NULL")# SE LIMPIA LA TABLA
tmp = tabla6.show()
```

```python
from pyspark.sql.functions import col
from pyspark.sql import functions as F
# SE DEFINE UNA FUNCION PARA SACAR EL PROMEDIO TOTAL DE TODAS LAS CALIFICACIONES Y SE REDONDEAN LAS CALIFICACIONES A 2 CIFRAS DECIMALES
notas = [col('avg(Introduccion)'),col('avg(Equipo)'),col('avg(Problema)'),col('avg(Ventajas)'),col('avg(Solucion)'),col('avg(Producto)'),col('avg(Traccion)'),col('avg(Mercado)'),col('avg(Competencia)'),col('avg(Negocio)'),col('avg(Inversion)'),col('avg(Contacto)'),col('avg(ExpoCoord)'),col('avg(ExpoExp)'),col('avg(ExpoDiapo)')]

# SE APLICA LA FUNCIÓN 
funpromedio = sum(x for x in notas)/len(notas)
tabla7 = tabla6.withColumn('PromedioTotal', funpromedio)

# SE REDONDEAN LAS CALIFICACIONES
for val in tabla7.columns:
    if val != 'EquipoEvaluar' and val != 'first(default.evaluacion.NombreUsuario)' :
        tabla7 = tabla7.withColumn(val, F.round(val,2))
tabla7.show()

```
